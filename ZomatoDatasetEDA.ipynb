```python
final_df = pd.merge(df, df_country, on='Country Code',how='left')
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python

plt.pie(country_values[:4], labels=country_names[:4],autopct='%1.2f%%')
```

Observations : zoamto maximizes transactions from India, then US then UK.

```python
ratings = final_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

 Observations
1. when ratings is between 4.5 to 4.9 -----> Excellent
2. when rating is between 4.0 to 4.4 -----> very good
3. when rating is between 3.5 to 3.9 -----> good
4. when rating is between 2.5 to 3.4  -----> Avg

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue = 'Rating color' ,data=ratings, palette=['grey','red','orange','yellow','green','green'])
```

observations
1.Not rated count is very high.
2.maximum number of observations of ratings are between 2.5 to 3.4.

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.countplot(x="Rating color",data=ratings, palette=['grey','red','orange','yellow','green','green'])
```
